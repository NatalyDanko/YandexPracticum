# Определяющие успешность игры закономерности

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as st

pd.set_option('display.max_columns', None)
```

```python
# загрузка данных

try:
    data = pd.read_csv('games.csv')    
except:
    print('Ошибка загрузки данных')
    
data.head()
```

```python
data.info()
```

# Предобработка данных

**Замена названий столбцов**

```python
data.columns = data.columns.str.lower()
```

**Проверка данных на явные дубликаты**

```python
data.duplicated().sum()
```

Явные дубликаты отсутствуют.

**Обработка пропусков**

```python
# проверка данных на пропуски

data.isna().sum()
```

```python
# так как пропусков в столбцах "name" и "genre" всего лишь два, удалим эти строки

data = data.dropna(subset=['name']).reset_index(drop=True)
data = data.dropna(subset=['genre']).reset_index(drop=True)

# оценим долю строк с пропусками в столбце "year_of_release" и в случае их небольшого количества также удалим их

ratio = data['year_of_release'].isna().sum() / data.shape[0]
print(f'{ratio:.1%}')
```

```python
# доля строк с пропусков  столбце "year_of_release" невелика, удалим эти строки

data = data.dropna(subset=['year_of_release']).reset_index(drop=True)
```

```python
# обработка значения "tdb" в столбце "user_score"

# данная аббревиатура расшифровывается, как "To Be Determined (Будет определено)", 
# поэтому было решено заменить эти значения на пропуски

data.loc[data['user_score'] == 'tbd','user_score'] = np.nan
```

**Обработка пропусков в столбце "rating"**

```python
# проверка доли строк с пропуском в столбце "rating"

ratio = data['rating'].isna().sum() / data.shape[0]
print(f'{ratio:.1%}')
```

Процент пропусков очень велик. Чтобы не потерять данные, заменим пропуски в этом столбце на "Без рейтинга".

```python
data['rating'] = data['rating'].fillna('Без рейтинга')
```

Заполнение пропусков в столбцах "critic_score" и "user_score" чем-либо логичным не представляется возможным.

```python
data.isna().sum()
```

**Преобразование типов данных в некоторых столбцах**

Смена типа данных в столбцах "year_of_release" и "user_score" c "float64" и "object" на более подходящие "int" и "float".

```python
data['year_of_release'] = data['year_of_release'].astype('int')
data['user_score'] = data['user_score'].astype('float')
```

```python
data.info()
```

**Добавление столбца суммарных продаж по всем регионам**

```python
data['sales_total'] = data[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)
data.head()
```

# Портрет пользователя каждого региона

**Самые популярные платформы по регионам**

```python
# подготовка сводной таблицы

data_platform_region =\
data.pivot_table(index='platform', values=['na_sales', 'eu_sales', 'jp_sales', 'sales_total'], aggfunc='sum')
data_platform_region
```

```python
# объявление функции для вывода круговой диаграммы

def print_platform_pie(dataset, region):
    
    popular_platforms = list(dataset.sort_values(by=region, ascending=False).index[:5])
    
    popular_platforms_sales = []
    
    for platform in popular_platforms:
        popular_platforms_sales.append(dataset.loc[platform, region])
    
    popular_platforms.append('Прочие')
    popular_platforms_sales.append(dataset[region].sum() - sum(popular_platforms_sales))

    fig1, ax = plt.subplots()
    colors = ['cornflowerblue', 'mediumaquamarine', 'plum', 'lightblue', 'bisque', 'lightgrey']
    ax.pie(popular_platforms_sales, labels=popular_platforms, autopct='%1.1f%%', colors=colors)
    ax.axis('equal')  
    fig = plt.gcf()
    fig.set_size_inches(7, 7)
    plt.show()
```

*Анализ региона "North America"*

```python
print_platform_pie(data_platform_region, 'na_sales')
```

*Анализ региона "Europe"*

```python
print_platform_pie(data_platform_region, 'eu_sales')
```

*Анализ региона "Japan"*

```python
print_platform_pie(data_platform_region, 'jp_sales')
```

На диаграммах видно, что в зависимости от региона предпочитаемые платформы отличаются друг от друга во всех трёх регионах.

# Проверка гипотез

Для проверки гипотез воспользуемся t-тестом - статистическим тестом с использованием t-распределения Стьюдента. Его выбор обусловлен тем, что если значений в выборке много (30 и больше), то распределение Стьюдента очень близко к нормальному, а также он работает для выборки любого размера.

**Проверка гипотезы: средние пользовательские рейтинги платформ Xbox One и PC одинаковы**

Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковы.  
Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC разные.

```python
# выборка необходимых массивов данных

sample_1 = data.query('platform == "XOne"')['user_score']
sample_2 = data.query('platform == "PC"')['user_score']

# критический уровень статистической значимости

alpha = 0.05

results = st.ttest_ind(sample_1, sample_2, equal_var = False, nan_policy='omit')

print('Значение p-value равно', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Проверка гипотезы: средние пользовательские рейтинги жанров Action и Sports разные**

Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports одинаковы.  
Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.

```python
# выборка необходимых массивов данных

sample_1 = data.query('genre == "Action"')['user_score']
sample_2 = data.query('genre == "Sports"')['user_score']

# критический уровень статистической значимости

alpha = 0.05

results = st.ttest_ind(sample_1, sample_2, equal_var = False, nan_policy='omit')

print('Значение p-value равно', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```
