***Обработка пропусков в столбце `ceiling_height`***

```python
sorted(data['ceiling_height'].sort_values().unique())
```

Обнаружены как слишком маленькие, так и слишком большие значения высоты потолков. Разделим значения большие, чем 10 м, на 10 (скорее всего в этих значениях неверно проставлена десятичная запятая) и снова сделаем проверку.

```python
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height'] < 11, data['ceiling_height'] / 10 )

sorted(data['ceiling_height'].sort_values().unique())
```

Выведем числовое описание данных столбца, чтобы понять, какие значения можно отсечь.

```python
data['ceiling_height'].describe()
```

Видим, что медиана близка к средней высоте потолков по СНиП. Стандартное отклонение невелико, значит данные более менее однородны, в размах «усов» входят значения от 2.1 м до 3.2 м. Минимальная высота потолков по СНиП должна быть не меньше 2.5 м, ограничимся этим значением, как минимальным: навряд ли есть двухуровневые квартиры, а если и есть, то про них следовало бы указать отдельно. Таким образом, подсчитаем, в какой доле объявлений высота потолков меньше 2.5 м и больше 3.2 м.

```python
ratio = data.query('ceiling_height < 2.5 or ceiling_height > 3.2')['ceiling_height'].count() / data.shape[0]
print(f'{ratio:.1%}')
```

Процент таких объявлений невелик, поэтому исключим из датафрейма соответствующие строки. Заполним пропуски медианным значением и проверим количество пропусков в столбце.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
data = data.query('ceiling_height >= 2.5 and ceiling_height <= 3.2')
data['ceiling_height'].isna().sum()
```

***Обработка пропусков в столбцах `living_area` и `kitchen_area`***

Для заполнения пропусков в данных столбцах сгруппируем объявления по количеству комнат и найдём средние значения для `living_area` и `kitchen_area`. Предварительно проверим значения для количества комнат `rooms` для студий и в случае необходимости заменим их на "0", чтобы отделить их от полноценных однокомнатных квартир: 

```python
# проверка уникальных значений количества комнат

data[data['studio'] == True]['rooms'].value_counts()
```

```python
# так как присутствуют два значения, 0 и 1, производим замену и делаем проверку

data.loc[data['studio'] == True, 'rooms'] = 0
data[data['studio'] == True]['rooms'].value_counts()
```

Также было замечено, что во всех студях отсутствуют данные для прощади кухни (равны NaN). В данном случае это не ошибки в данных, а специфика такого типа квартир, поэтому удалять строки с пропуском в kitchen_area не будем, чтобы не удалить все студии из датафрейма.

```python
data[data['studio'] == True]['kitchen_area'].isna().sum()
```

Кроме того, датафрейме присутствуют квартиры, в которых количество комнат равно нулю, но которые не являются студиями. Выведем числовое описание данных столбцов `total_area`, `living_area` и `kitchen_area` для таких квартир и для студий, чтобы понять, можно ли будет объединить эти квартиры при группировке по количеству квартир:

```python
# числовое описание для студий 
print(data.query('studio == True')[['total_area', 'living_area', 'kitchen_area']].describe(), end='\n\n')

# числовое описание для квартир, с количеством комнат, равным нулю, не являющихся студиями
print(data.query('studio == False and rooms == 0')[['total_area', 'living_area', 'kitchen_area']].describe())
```

Видно, что наборы данных описывают примерно одни и те же цифры, особенно для межквартильного размаха. Поэтому сделаем группировку по количеству комнат.

```python
# находим медианные значения для 'living_area' и 'kitchen_area' для объявлений, сгруппированных по количеству комнат

median_areas = data.pivot_table(index='rooms', values=['living_area', 'kitchen_area'], aggfunc=['median'])
median_areas.columns = ['kitchen_area_median', 'living_area_median']
print(median_areas)
```

```python
# заполняем пропуски средними значениями 

for room in data['rooms'].unique():
    data.loc[data['rooms'] == room, 'kitchen_area'] = \
    data.loc[data['rooms'] == room, 'kitchen_area'].fillna(median_areas.loc[room, 'kitchen_area_median'])
    
    data.loc[data['rooms'] == room, 'living_area'] = \
    data.loc[data['rooms'] == room, 'living_area'].fillna(median_areas.loc[room, 'living_area_median'])
```

```python
# проверим пропуски

data[['living_area', 'kitchen_area']].isna().sum()
```

```python
# оставшиеся пропуски - это квартиры студии и квартиры с количеством квартир, равным нулю, но студиями не являющимися

print(data[data['studio'] == True]['rooms'].count(), data.query('studio == False and rooms == 0')['rooms'] .count(), sep='\n')
```

***Обработка пропусков в столбцах `airports_nearest` и `cityCenters_nearest`*** 

Для заполнения пропусков в данных столбцах сгруппируем объявления по расположению и найдём медианные значения для `airports_nearest` и `cityCenters_nearest`, где это возможно.

### Добавление в таблицу новых столбцов

```python
# расчёт стоимости одного квадратного метра

data['cost_per_square_meter'] = data['last_price'] / data['total_area']

# добавление столбцов с днём, месяцем и годом публикации объявления

data['day_of_exposition'] = data['first_day_exposition'].dt.weekday
data['month_of_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['year_of_exposition'] = pd.DatetimeIndex(data['first_day_exposition']).year
```

```python
# добавление столбца с указанием типа этажа: первый, последний, другой

def get_floor_type(row):
    
    floors_total = row['floors_total']
    floor = row['floor']
    
    if floor == floors_total:
        return 'последний'
    
    elif floor == 1:
        return 'первый'
    
    else:
        return 'другой'
    
data['floor_type'] = data.apply(get_floor_type, axis=1)
```

```python
# добавление расстояния в км до центра города

data['cityCenters_distance_km'] = data['cityCenters_nearest'] / 1000

data.head()
```

**Нахождение максимальной и минимальной средней цены одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений.**

```python
# города с максимальным количеством объявлений 

data['locality_name'].value_counts().head(10)
```

```python
loc_names = ['Санкт-Петербург', 'поселок Мурино', 'поселок Шушары',\
             'Всеволожск', 'Колпино', 'Пушкин', 'Гатчина', 'деревня Кудрово', 'поселок Парголово', 'Выборг']

data_loc_names = data.query('locality_name in @loc_names')
avg_price = data_loc_names.pivot_table(index='locality_name', values=['last_price', 'total_area'], aggfunc='sum')

# добавление столбца со средней ценой одного квадратного метра

avg_price['avg_price_per_meter'] = avg_price['last_price'] / avg_price['total_area']
avg_price.sort_values(by='avg_price_per_meter')
```

Таким образом, самая большая цена за один квадратный метр в Санкт-Петербурге: около 104 тысяч. Самая  маленькая цена за один квадратный метр в Выборге: около 58 тысяч.

**Вычисление средней цены каждого километра расстояния от центра города для квартир из Санкт-Петербурга.**

```python
import numpy as np

data_peter = data.query('locality_name == "Санкт-Петербург"')[['last_price', 'total_area', 'cityCenters_distance_km']]
data_peter['avg_price_per_meter'] = data['last_price'] / data['total_area']
data_peter['cityCenters_distance_km'] = np.round(data_peter['cityCenters_distance_km']).astype('int')
```

```python
data_peter.pivot_table(index='cityCenters_distance_km', values='avg_price_per_meter')
```

```python
print('Коэффициент кореляции Пирсона для зависимости стоимости объектов от расстояния до центра города:', \
     data_peter['last_price'].corr(data_peter['cityCenters_distance_km']))
```

Слабая корреляция с отрицательным значением, что означает, что стоимость квартир увеличивается с уменьшением расстояния от центра города.
